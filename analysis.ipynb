```python
import pandas as pd
df = pd.read_csv("tops_crime_data_cleaned.csv")
```

```python
crime_counts = df["DISPO_TEXT"].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
crime_counts.plot(kind="bar")

plt.title("Top 10 Most Common Crimes")
plt.xlabel("Crime Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()

# save the plot as an image file
plt.savefig("figures/top_10_crimes.png", dpi=300)

plt.show()
```
